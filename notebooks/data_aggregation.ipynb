```python
import pandas as pd
```

```python
# drop with columns with zero
```

```python
df_2017 = pd.read_csv("2017_full_data.csv").drop('Unnamed: 0', axis=1)
df_2018 = pd.read_csv("2018_full_data.csv").drop('Unnamed: 0', axis=1)
df_2019 = pd.read_csv("2019_full_data.csv").drop('Unnamed: 0', axis=1)
df_2020 = pd.read_csv("2020_full_data.csv").drop('Unnamed: 0', axis=1)
df_2021 = pd.read_csv("2021_full_data.csv").drop('Unnamed: 0', axis=1)
```

```python
df = pd.concat([df_2017, df_2018, df_2019, df_2020, df_2021])
```

```python
data = df.drop(['Location ID', 'Location Name',
             'Location Type', 'Energy Component', 'Congestion Component', 
             'Marginal Loss Component', 'lag25', 'lag26', 'lag27', 'lag28',
       'lag29', 'lag30', 'lag31', 'lag32', 'lag33', 'lag34', 'lag35', 'lag36',
       'lag37', 'lag38', 'lag39', 'lag40', 'lag41', 'lag42', 'lag43', 'lag44',
       'lag45', 'lag46', 'lag47'], axis=1)
```

```python
# Set date time column.

data = data[data['Hour_Ending'] != "02X"]

data['Hour_Ending'] = data['Hour_Ending'].astype(int)
data['Hour_Ending'] = data['Hour_Ending'] - 1
data['Date'] = pd.to_datetime(data['Date']) + pd.to_timedelta(data['Hour_Ending'], unit='h')
data = data.drop(['Hour_Ending'], axis=1)
```
